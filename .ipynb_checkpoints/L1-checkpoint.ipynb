```python
import torch
import transformer_lens

print("CUDA available:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))

model = transformer_lens.HookedTransformer.from_pretrained("gpt2-medium")
model = model.to("cuda")
print("Model loaded on GPU")
print("Parameters:", sum(p.numel() for p in model.parameters()) / 1e6, "M")
```

```python
tokens = model.to_tokens("Hyderabad is in the state of")
print(tokens.shape)
print(tokens)
print(model.to_str_tokens("Hyderabad is in the state of"))
logits,cache = model.run_with_cache(tokens)
for key in list(cache.keys()):
    print(key,"->",cache[key].shape)
```

```python
# Get the top 5 predictions for the next token
import torch

next_token_logits = logits[0, -1, :]  # last token position
top5 = torch.topk(next_token_logits, 5)

print("GPT-2 thinks 'Hyderabad is in the state of ___':")
print()
for prob, idx in zip(top5.values, top5.indices):
    token = model.tokenizer.decode([idx])
    print(f"  '{token}' — score: {prob:.2f}")
```

```python
# Extract the residual stream at every layer for the LAST token position
# This is what we'll use to train our hallucination probe

print("Residual stream at last token ('of') across all layers:")
print()
residual_snapshots = []

for layer in range(model.cfg.n_layers):
    resid = cache["resid_post", layer]  # [batch, seq, d_model]
    last_token_resid = resid[0, -1, :]  # just the last token
    residual_snapshots.append(last_token_resid)
    print(f"  Layer {layer}: shape {last_token_resid.shape}, "
          f"mean {last_token_resid.mean().item():.4f}, "
          f"std {last_token_resid.std().item():.4f}")

print(f"\nTotal layers: {model.cfg.n_layers}")
print(f"d_model (hidden size): {model.cfg.d_model}")
print(f"\nThis gives us {model.cfg.n_layers} vectors of size {model.cfg.d_model}")
print(f"Each vector = one potential feature for our hallucination classifier")
```

```python
# Properly labeled dataset — label is determined by 
# comparing GPT-2's actual prediction to the true answer

examples = [
    # (prompt, true_answer_fragment)
    ("Paris is the capital of", "France"),
    ("The Sun is at the center of the solar", "system"),
    ("Water boils at 100 degrees", "Celsius"),
    ("The Great Wall is located in", "China"),
    ("Mount Everest is the tallest mountain in the", "world"),
    ("Hyderabad is in the state of", "Telangana"),
    ("The capital of Australia is", "Canberra"),
    ("Einstein won the Nobel Prize for his discovery of the", "photoelectric"),
    ("The inventor of the telephone was Alexander Graham", "Bell"),
    ("Pluto is classified as a dwarf", "planet"),
]

dataset = []

print("Building dataset — checking GPT-2 predictions vs truth:\n")

for prompt, true_answer in examples:
    tokens = model.to_tokens(prompt)
    logits = model(tokens)
    
    # Get top prediction
    next_logits = logits[0, -1, :]
    top1_token = model.tokenizer.decode([next_logits.argmax().item()]).strip().lower()
    true_lower = true_answer.lower()
    
    # Label: 1 if GPT-2 is WRONG (hallucination), 0 if correct
    is_hallucination = 0 if true_lower in top1_token or top1_token in true_lower else 1
    
    # Extract residual stream at layer 22 (where hallucination signal is strongest)
    _, cache = model.run_with_cache(tokens)
    activation = cache["resid_post", 22][0, -1, :].cpu().detach().numpy()
    
    dataset.append({
        "prompt": prompt,
        "true_answer": true_answer,
        "gpt2_predicted": top1_token,
        "label": is_hallucination,
        "activation": activation
    })
    
    status = "HALLUCINATION" if is_hallucination else "CORRECT"
    correct_mark = "✗" if is_hallucination else "✓"
    print(f"  {correct_mark} [{status:13}] '{prompt}'")
    print(f"    True: '{true_answer}' | GPT-2 said: '{top1_token}'")
    print()

hallucination_count = sum(d["label"] for d in dataset)
print(f"Dataset: {len(dataset)} examples — {hallucination_count} hallucinations, "
      f"{len(dataset)-hallucination_count} correct")
```
